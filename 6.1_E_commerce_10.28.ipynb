# 清理数据
假设：代做清单

根据前面评估部分得到的结论,我们需要进行的数据清理包括:

· 把InvoiceDate变量的数据类型转换为为日期时间

· 把CustomerID变量的数据类型转换为字符串

· 把Description变量缺失的观察值删除

· 把Country 变量值“USA”替换为"United States"

· 把Country 变量值“UK”、“U.K.”替换为"United Kingdom"

· 把Quantity变量值为负数的观察值删除

■ 把UnitPrice 变量值为负数的观察值删除

为了区分开经过清理的数据和原始的数据,我们创建新的变量cleaned_data,让它为original_data复制出的副本。我们
之后的清理步骤都将被运用在cleaned_data上。

```python
cleaned_data = original_data.copy()
cleaned_data.head()
```

把 InvoiceDate 变量的数据类型转换为为日期时间类型:

```python
cleaned_data["Browsing Time (min)"] = pd.to_datetime(cleaned_data["Browsing Time (min)"])
cleaned_data["Browsing Time (min)"]
```

把Discount 变量的数据类型转换为字符串:

```python
cleaned_data["Discount"] = cleaned_data["Discount"].astype(str)
cleaned_data["Discount"]
```

把Discount 变量值结尾的“.0”删除:

```python
cleaned_data["Discount"]=cleaned_data["Discount"].str.slice(0,-2)
cleaned_data["Discount"]
```

把Order Priority 变量缺失的观察值删除,并查看删除后该列空缺值个数和:

```python
# ​如果是 Series​（一维数据）：
# •cleaned_data = cleaned_data.dropna(subset=["Order Priority"])
# cleaned_data["Order Priority"].isnull().sum()
# Series的 dropna()​不支持 subset，因为它只能对整个 Series 删除缺失值。
cleaned_data = pd.DataFrame(cleaned_data)  # 转换为DataFrame
cleaned_data = cleaned_data.dropna(subset=["Order Priority"])
cleaned_data["Order Priority"].isnull().sum()
```
